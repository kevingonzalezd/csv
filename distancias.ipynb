```python
import pandas as pd
import googlemaps
df = pd.read_csv('direcciones.csv')
df = df[df['Calle'] != '0']
df['numero de casa'] =df['numero de casa'].astype(int) 
df['direccion_completa'] = df['Calle'] + ', ' + df['numero de casa'].astype(str) + ', ' + df['Colonia'] + ', ' + df['ciudad'] + ', ' + 'Estado Ejemplo' + ', ' + df['codigo postal'].astype(str) + ', ' + 'México'
df
```

```python
gmaps = googlemaps.Client(key='AIzaSyARwNqgineqz--YKiUeJjKmu7RCWEdf3nY')

df=primeras_10_filas = df.head(50)

# Crea una función para obtener la distancia entre dos direcciones.
def obtener_distancia(origen, destino):
    try:
        distancia = gmaps.distance_matrix(origen, destino)["rows"][0]["elements"][0]
        return distancia["distance"]["value"] / 1000  # La distancia en kilómetros
    except Exception as e:
        print(f"Error al obtener distancia: {e}")
        return None
    
# Crea una matriz de distancias vacía
num_direcciones = len(df)
matriz_distancias = [[None] * num_direcciones for _ in range(num_direcciones)]

# Llena la matriz de distancias
for i in range(num_direcciones):
    for j in range(num_direcciones):
        if i != j:
            origen = df['direccion_completa'][i]
            destino = df['direccion_completa'][j]
            distancia = obtener_distancia(origen, destino)
            matriz_distancias[i][j] = distancia

# Convierte la matriz de distancias en un DataFrame
df_distancias = pd.DataFrame(matriz_distancias)
df_distancias
# Guarda el DataFrame de distancias en un archivo si es necesario
#df_distancias.to_csv('distancias.csv', index=False)
```
