```python
import random
import math
import time
import pandas as pd

def dimension1(pasos):
    posicion = 0
    distancia_maxima = 0
    for _ in range(pasos):
        step = random.choice([-1, 1])
        posicion += step
        distancia_maxima = max(distancia_maxima, abs(posicion))
    distance_to_origin = abs(posicion)
    return distancia_maxima, distance_to_origin

def dimension2(pasos):
    x, y = 0, 0
    distancia_maxima = 0
    for _ in range(pasos):
        direccion = random.choice([(1, 0), (-1, 0), (0, 1), (0, -1)])
        x_paso, y_paso = direccion
        x += x_paso
        y += y_paso
        distance = math.sqrt(x**2 + y**2)
        distancia_maxima = max(distancia_maxima, distance)
    distance_to_origin = math.sqrt(x**2 + y**2)
    return distancia_maxima, distance_to_origin

def simulacion(dimensions, pasos, reps):
    distancia_maximas = []
    distances_to_origin = []
    results = {'Repeticion': [], 'Distancia Maxima': [], 'Distancia al Origen': []}
    start_time = time.time()
    for rep in range(reps):
        if dimensions == 1:
            distancia_maxima, distance_to_origin = dimension1(pasos)
        elif dimensions == 2:
            distancia_maxima, distance_to_origin = dimension2(pasos)
        distancia_maximas.append(distancia_maxima)
        distances_to_origin.append(distance_to_origin)
        results['Repeticion'].append(rep + 1)
        results['Distancia Maxima'].append(distancia_maxima)
        results['Distancia al Origen'].append(distance_to_origin)
    end_time = time.time()
    tiempo_ejecucion = end_time - start_time
    return distancia_maximas, distances_to_origin, tiempo_ejecucion, results

dimensions = int(input("Ingrese el número de dimensiones (1 o 2): "))
pasos = int(input("Ingrese el número de pasos por caminata: "))
reps = int(input("Ingrese la cantidad de repeticiones: "))

distancia_maximas, distances_to_origin, tiempo_ejecucion, results = simulacion(dimensions, pasos, reps)

distancia_promedio = sum(distancia_maximas) / reps
distancia_promedio_to_origin = sum(distances_to_origin) / reps

data = {'Dimensiones': [dimensions],
        'Pasos': [pasos],
        'Repeticiones': [reps],
        'Promedio Distancia Máxima': [distancia_promedio],
        'Promedio Distancia al Origen': [distancia_promedio_to_origin],
        'Tiempo de Ejecución (s)': [tiempo_ejecucion]}

df = pd.DataFrame(data)
df
```
