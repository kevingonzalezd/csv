# Algoritmo con solución Factible

```python
import time
inicio = time.time()
import random

contenedores = [("C1",61,1100),
                ("C2",58,1147),
                ("C3",92,1442),
                ("C4",50,1591),
                ("C5",108,1078),
                ("C6",83,1385),
                ("C7",93,1777),
                ("C8",101,1196),
                ("C9",54,1753),
                ("C10",50,1371),
                ("C11",72,1517),
                ("C12",51,1675),
                ("C13",100,1193),
                ("C14",108,1177),
                ("C15",91,1365),
                ("C16",112,1143),
                ("C17",66,1314),
                ("C18",58,1526),
                ("C19",110,1470),
                ("C20",73,1605),
                ]

# Capacidad Carguero
capacidad = 800

# Parámetros
tamanoPob = 20
probMutacion = 0.03
generaciones = 10
criterioParo = 10
probCruce = 0.9

print("\nParametros:")
print("Tamaño de Población:", tamanoPob)
print("Probabilidad de Mutación:", probMutacion)
print("Generaciones máximas:", generaciones, "\n")

# Función Fitness
def fitness(cromosoma):
    pesoTotal = 0
    valorTotal = 0
    for i in range(len(cromosoma)):
        if cromosoma[i] == 1:
            pesoTotal += contenedores[i][1]
            valorTotal += contenedores[i][2]*contenedores[i][1]

    return pesoTotal, valorTotal

# Generacion de Población
poblacion = []
for i in range(tamanoPob):
    cromosoma = [0] * len(contenedores)
    pesoActual = 0
    valorActual = 0
    intentos = 0
    while pesoActual < capacidad and intentos < len(contenedores):
        indice = random.randint(0, len(contenedores) - 1)
        if cromosoma[indice] == 0 and pesoActual + contenedores[indice][1] <= capacidad:
            cromosoma[indice] = 1
            pesoActual += contenedores[indice][1]
            valorActual += contenedores[indice][2]
        intentos += 1
    poblacion.append(cromosoma)

# Algoritmo genético
generacion = 0
while generacion < generaciones:
    # Aptitud de cada cromosoma
    aptitudes = [fitness(cromosoma) for cromosoma in poblacion]

    nuevaPob = []
    for i in range(tamanoPob):
        padre1 = random.choices(poblacion, weights=[apt[0] for apt in aptitudes])[0]
        padre2 = random.choices(poblacion, weights=[apt[0] for apt in aptitudes])[0]

        # Operador de cruce con probabilidad probCruce
        if random.random() < probCruce:
            puntoCruce = random.randint(1, len(contenedores) - 1)
            hijo1 = padre1[:puntoCruce] + padre2[puntoCruce:]
            hijo2 = padre2[:puntoCruce] + padre1[puntoCruce:]
        else:
            hijo1 = padre1[:]
            hijo2 = padre2[:]

        # Operador de mutación con probabilidad probMutacion
        for i in range(len(contenedores)):
            if random.random() < probMutacion:
                hijo1[i] = 1 - hijo1[i]
                hijo2[i] = 1 - hijo2[i]

        # Limitar el peso de los hijos
        pesoHijo1, _ = fitness(hijo1)
        pesoHijo2, _ = fitness(hijo2)
        if pesoHijo1 > capacidad:
            for j in range(len(contenedores)):
                if hijo1[j] == 1 and pesoHijo1 > capacidad:
                    hijo1[j] = 0
                    pesoHijo1 -= contenedores[j][1]
        if pesoHijo2 > capacidad:
            for j in range(len(contenedores)):
                if hijo2[j] == 1 and pesoHijo2 > capacidad:
                    hijo2[j] = 0
                    pesoHijo2 -= contenedores[j][1]

        # Juntamos hijos nuevos en población nueva
        nuevaPob.append(hijo1)
        nuevaPob.append(hijo2)

    # Actualizamos población y contador de generaciones
    poblacion = nuevaPob
    generacion += 1

# Encontrar el mejor cromosoma en la población final
mejorCromosoma = max(poblacion, key=lambda cromosoma: fitness(cromosoma)[1])
pesoMejorCromosoma, valorMejorCromosoma = fitness(mejorCromosoma)
final = time.time()
tiempo = final - inicio

# Resultados
print("Solución factible encontrada:")
print("Cromosoma:", mejorCromosoma)
print("Valor: $", valorMejorCromosoma)
print("Peso Total:", pesoMejorCromosoma)
print("Tiempo de Ejecución:", tiempo)
```

```python
import time
inicio = time.time()
import random

#Objetos para meter en carguero
contenedores = [("C1",61,1100),
                ("C2",58,1147),
                ("C3",92,1442),
                ("C4",50,1591),
                ("C5",108,1078),
                ("C6",83,1385),
                ("C7",93,1777),
                ("C8",101,1196),
                ("C9",54,1753),
                ("C10",50,1371),
                ("C11",72,1517),
                ("C12",51,1675),
                ("C13",100,1193),
                ("C14",108,1177),
                ("C15",91,1365),
                ("C16",112,1143),
                ("C17",66,1314),
                ("C18",58,1526),
                ("C19",110,1470),
                ("C20",73,1605),
                ]


# Capacidad Carguero
capacidad = 800

# Parámetros

tamanoPob = 20
probMutacion = 0.03
generaciones = 10
criterioParo = 10
probCruce = 0.9


print("\nParametros:")
print("Tamaño de Población:", tamanoPob)
print("Probabilidad de Mutación:", probMutacion)
print("Generaciones:", generaciones, "\n")


# Función Fitness
def fitness(cromosoma):
    pesoTotal = 0
    valorTotal = 0
    for i in range(len(cromosoma)):
        if cromosoma[i] == 1:
            pesoTotal += contenedores[i][1]
            valorTotal += contenedores[i][2] * contenedores[i][1]

    if pesoTotal > capacidad:
        return valorTotal, pesoTotal
    else:
        return valorTotal, 0


# Población aleatoria
poblacion = []
for i in range(tamanoPob):
    cromosoma = [random.randint(0, 1) for i in range(len(contenedores))]
    poblacion.append(cromosoma)

# Algoritmo genético
generaciones = 0
while generaciones < criterioParo:

    # Aptitud de cada cromosoma
    aptitudes = [fitness(cromosoma) for cromosoma in poblacion]

    #Generamos población
    nuevaPob = []
    for i in range(tamanoPob):
        padre1 = random.choices(poblacion, weights=[apt[0] for apt in aptitudes])[0]
        padre2 = random.choices(poblacion, weights=[apt[0] for apt in aptitudes])[0]

        # Operador de cruce con probabilidad probCruce
        if random.random() < probCruce:
            puntoCruce = random.randint(1, len(contenedores) - 1)
            hijo1 = padre1[:puntoCruce] + padre2[puntoCruce:]
            hijo2 = padre2[:puntoCruce] + padre1[puntoCruce:]
        else:
            hijo1 = padre1[:]
            hijo2 = padre2[:]

        # Operador de mutación con probabilidad probMutacion
        for i in range(len(contenedores)):
            if random.random() < probMutacion:
                hijo1[i] = 1 - hijo1[i]
                hijo2[i] = 1 - hijo2[i]
        # Juntamos hijos nuevos en población nueva
        nuevaPob.append(hijo1)
        nuevaPob.append(hijo2)

    # Actualizamos población y agregamos al contador de gen
    poblacion = nuevaPob
    generaciones += 1

# Encontrar el mejor cromosoma en la población final
mejorCromosoma = max(poblacion, key=lambda cromosoma: fitness(cromosoma)[0])
mejorValor, pesoTotal = fitness(mejorCromosoma)
final = time.time()
tiempo = final - inicio
print("Mejor solución encontrada:")
print("Cromosoma:", mejorCromosoma)
print("Valor:", mejorValor)
print("Peso Total:", pesoTotal)
print("Tiempo de Ejecución:", tiempo )
```
